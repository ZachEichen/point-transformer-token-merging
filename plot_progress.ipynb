```python
import matplotlib.pyplot as plt
import json
```

```python
with open("progress_dict.json", mode="r") as f:
    progress_dict = json.load(f)
```

```python
step_keys = list(progress_dict.keys())

train_steps = [int(s.split("_")[1]) for s in step_keys]
train_loss = [progress_dict[s]["train"]["loss"] for s in step_keys]
train_acc = [progress_dict[s]["train"]["accuracy"] for s in step_keys]
train_f1 = [progress_dict[s]["train"]["f1"] for s in step_keys]

val_steps = [int(s.split("_")[1]) for s in step_keys if "val" in progress_dict[s]]
val_loss = [progress_dict[s]["val"]["loss"] for s in step_keys if "val" in progress_dict[s]]
val_acc = [progress_dict[s]["val"]["accuracy"] for s in step_keys if "val" in progress_dict[s]]
val_f1 = [progress_dict[s]["val"]["f1"] for s in step_keys if "val" in progress_dict[s]]
```

```python
plt.plot(train_steps, train_loss, label='Training Loss')
plt.xlabel('Steps')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
plt.plot(val_steps, val_loss, label='Validation Loss')
plt.xlabel('Steps')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
plt.plot(train_steps, train_acc, label='Training Acc')
plt.plot(val_steps, val_acc, label='Validation Acc')
plt.xlabel('Steps')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
plt.plot(train_steps, train_f1, label='Training F1')
plt.plot(val_steps, val_f1, label='Validation F1')
plt.xlabel('Steps')
plt.ylabel('F1')
plt.legend()
plt.show()
```
