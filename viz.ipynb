```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("outputs/test/1/progress.csv")
df.columns
```

```python
train_loss = list(df["T_Loss"])
val_loss = list(df["V_Loss"])
train_acc = list(df["T_Accuracy"])
val_acc = list(df["V_Accuracy"])
steps = list(df["step"])
```

```python
plt.plot(steps, train_loss, label="train_loss")
plt.plot(steps, train_loss, label="val_loss")
plt.legend()
plt.xlabel("steps")
plt.ylabel("loss")
plt.show()
```

```python
plt.plot(steps, train_acc, label="train_acc")
plt.plot(steps, val_acc, label="val_acc")
plt.legend()
plt.xlabel("steps")
plt.ylabel("accuracy")
plt.show()
```
