```python
progress = pd.read_csv("outputs/point_transformer/progress.csv")
# progress = pd.read_csv("outputs/pt_merged/progress.csv") 
progress
```

```python
train_steps = list(progress["step"])
train_loss = list(progress["T_Loss"])
train_acc = list(progress["T_Accuracy"])
train_f1 = list(progress["T_Precision"])
train_f1 = list(progress["T_Recall"])
```

```python
val_steps = list(progress[progress['V_Loss'].notna()]['step'])
val_loss = list(progress["V_Loss"][val_steps])
val_acc = list(progress["V_Accuracy"][val_steps])
val_f1 = list(progress["V_Precision"][val_steps])
val_f1 = list(progress["V_Recall"][val_steps])

print(val_acc)
```

```python
plt.plot(train_steps, train_loss, label='Training Loss')
plt.xlabel('Steps')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
plt.plot(val_steps, val_loss, label='Validation Loss')
plt.xlabel('Steps')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
plt.plot(train_steps, train_acc, label='Acc')
# plt.plot(val_steps, val_acc, label='Validation Acc')
plt.xlabel('Steps')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
plt.plot(train_steps, train_f1, label='Training F1')
plt.plot(val_steps, val_f1, label='Validation F1')
plt.xlabel('Steps')
plt.ylabel('F1')
plt.legend()
plt.show()
```
